```python
import matplotlib.pyplot as plt
```

```python
def traerdatos(x):
    listaprincipal = []

    with open(x) as archivo:
        for linea in archivo:
            listaprincipal.append(float(linea))
    return listaprincipal
```

```python

cy1 = traerdatos("tiemposejecucionbubble/tiemposc++.txt")

print(cy1)
cx1=[100,1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000,
    9000, 10000, 20000, 30000, 40000, 50000]
plt.plot(cx1,cy1,marker='o', linestyle='-', color = 'b',label='C++')
plt.legend()
plt.ylabel("Tiempo Promedio despues de 5 iteraciones")
plt.xlabel("Tama√±o del array")
plt.title("Tiempo Promedio c++")
plt.show()
```
